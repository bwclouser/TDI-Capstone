```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[10,6]

seed=2019
np.random.seed(seed)
```

```python
y0=0.2
m0=2
m1=3
m2=1
noise_var=0.1
num_samples=50000
x0 = np.arange(0,10,2e-4)
x1 = np.sin(2.*3.14159*x0*5)
x2 = np.sin(2.*3.14159*x0*25)

y = y0+m0*x0 + m1*x1 + +m2*x2+ np.random.normal(0, noise_var, num_samples)
linear_data_df = pd.DataFrame(data=np.column_stack([x0, x1, x2 ,y]), columns=['x0','x1','x2','y'])
linear_data_df.head()
```

```python
plt.plot(linear_data_df['x0'],linear_data_df['y'],'o', markersize=1, label='Generated data points', color='skyblue');
plt.plot(linear_data_df['x0'], y0 + m0*linear_data_df['x0']+m1*linear_data_df['x1'], 'r-', label='Noiseless linear function');
plt.legend()
```

```python
#from pyspark.sql import SparkSession
#sc=SparkSession.builder.master('local[*]').getOrCreate()
from pyspark import SparkContext
sc = SparkContext('local[*]','temp')
```

```python
from pyspark.sql import SQLContext
sqlContext=SQLContext(sc)
```

```python
data_df = sqlContext.createDataFrame(data=linear_data_df, schema=list(linear_data_df))
```

```python
data_df.printSchema()
data_df.show(5)
```

```python
from pyspark.ml.regression import LinearRegression
lr = LinearRegression()
```

```python
from pyspark.ml.feature import VectorAssembler

va = VectorAssembler(inputCols = ['x0','x1','x2'], outputCol = 'features')
vdata_df = va.transform(data_df)
vdata_df.show()
```

```python
prepped_df = vdata_df.selectExpr('features','y as label')
prepped_df.printSchema()
```

```python
lr_model=lr.fit(prepped_df)
print('Coefficients: {}'.format(lr_model.coefficients))
print('Intercept: {}'.format(lr_model.intercept))
```

```python
ts=lr_model.summary
```
